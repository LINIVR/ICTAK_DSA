1. Find out how many males and females participated in the test.

2. What do you think about the students' parental level of education?

```python
data.groupby(["parental level of education"]).describe()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x ='parental level of education', data=data)
plt.title('PARENTAL LEVEL OF EDUCATION COUNT',size=15)
```

```python
mean_maths = data.groupby("parental level of education")["math score"].mean().reset_index()
```

```python
mean_maths.sort_values("math score",ascending=False)
```

```python
mean_reading = data.groupby("parental level of education")["reading score"].mean().reset_index()
mean_reading.sort_values("reading score",ascending=False)
```

```python
mean_writing = data.groupby("parental level of education")["writing score"].mean().reset_index()
mean_writing.sort_values("writing score",ascending=False)
```

3. Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation course

1.Gender

```python
mean_maths_gender = data.groupby("gender")["math score"].mean().reset_index()
mean_maths_gender.sort_values("math score",ascending=False)
```

```python
mean_reading_gender = data.groupby("gender")["reading score"].mean().reset_index()
mean_reading_gender.sort_values("reading score",ascending=False)
```

```python
mean_writing_gender =data.groupby("gender")["writing score"].mean().reset_index()
mean_writing_gender.sort_values("writing score",ascending=False)
```

2.Test preparation course

```python
data.groupby("test preparation course")[["math score","reading score","writing score"]].mean().reset_index()
```
