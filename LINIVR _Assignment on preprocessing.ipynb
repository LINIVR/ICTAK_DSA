# 5.Check and handle outliers in at least 3 columns in the dataset

```python
   num_cols = ['Survived', 'Pclass','Age', 'SibSp', 'Parch', 'Fare']                                   #  lt.boxplot(df['Fare','Age','SibSp'])
```

```python
# 1.Handling the outliers of 'Fare'column
```

```python
Q1 = np.percentile(df['Fare'],25,interpolation='midpoint')
Q2 = np.percentile(df['Fare'],50,interpolation='midpoint')
Q3 = np.percentile(df['Fare'],75,interpolation='midpoint')
```

```python
IQR = Q3-Q1
low_lim = Q1-1.5*IQR
up_lim  = Q3+1.5*IQR
print(low_lim)
print(up_lim)
```

```python
outlier=[]
for x in df['Fare']:
    if((x>up_lim) or (x<low_lim)):
        outlier.append(x)
      
```

Flooring and Capping for handling outliers

```python
df['Fare']=np.where(df['Fare']>up_lim,up_lim,np.where(df['Fare']<low_lim,low_lim,df['Fare']))
```

```python
df
```

```python
plt.boxplot(df['Fare'])
plt.title('Fare')
plt.show()
```
